Without limiting the generality we will compare two ways of differentiation on gradients with respect to $V$.

````python
import numpy as np
import jax.numpy as jnp
from jax import grad, jit, vmap
import timeit

def custom_grad_fn(X, U, V):
    return -(U.T @ X - (U.T @ U) @ V)

def target(X, U, V):
    return 0.5 * jnp.linalg.norm(X - U @ V, ord='fro')**2

jax_grad_fn = grad(target, argnums=2)
````

# Problem 3. Compression of the fully-connected layers in neural network with simple architecture (30 pts)

In this problem we consider the neural network that performs classification of the dataset of images. 
Any neural network can be considered as composition of simple linear and non-linear functions.
For example, a neural network with 3 layers can be represented as 

$$f_3(f_2(f_1(x, w_1), w_2), w_3),$$

where $x$ is input data (in our case it will be images) and $w_i, \; i =1,\dots,3$ are parameters that are going to be trained. 

We will study the compression potential of neural network with simple architecture: alternating some numbers of linear and non-linear functions. 

The main task in this problem is to study how the compression of fully-connected layers affects the test accuracy.
Any fully-connected layer is represented as linear function $AX + B$, where $X$ is input matrix and $A, B$ are trainable matrices. Matrices $A$ in every layer are going to be compressed.
The main result that you should get is the plot of dependence of test accuracy on the total number of parameters in the neural network.

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader, verbose=True):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    if verbose:
        print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
            test_loss, correct, len(test_loader.dataset),
            100. * correct / len(test_loader.dataset)))
    return 100. * correct / len(test_loader.dataset)
````

### Set parameters for training and print intermediate loss values

````python
log_interval = 50
epochs = 7
````

### Third step: run training with the [Adam](https://arxiv.org/pdf/1412.6980.pdf%20%22%20entire%20document) optimization method

If your laptop is not very fast, you will wait some time till training is finished. 

````python
model = Net()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(1, epochs + 1):
    train(model,  train_loader, optimizer, epoch)
    test(model, test_loader)
````

````python
from collections import OrderedDict
import re

class CompressedNet(nn.Module):
    def __init__(self, net: Net, rank: int):
        super().__init__()
        self.fc1 = net.fc1
        self.fc6 = net.fc6
        self.ReLU = net.ReLU
        self.rank = rank

        self.layers = OrderedDict()
        wait_for_bias = False
        u, s, v = None, None, None
        for n, p in net.named_parameters():
             if p.ndim == 2 and p.shape[0] == 1000 and p.shape[1] == 1000:
                u, s, v = torch.svd(p.data)
                u, s, v = nn.Parameter(u[:, :rank]), nn.Parameter(s[:rank]), nn.Parameter(v[:, :rank])
                wait_for_bias = True
             elif p.ndim == 1 and p.shape[0] == 1000 and wait_for_bias:
                self.layers[re.sub(r'.bias', '', n)] = (u, s, v, nn.Parameter(p.data))
                wait_for_bias = False

    @staticmethod
    def linear_forward(u, s, v, b, x):
        x_ = s[:, None] * (v.T @ x.T)
        x_ = u @ x_
        return x_.T + b[None, :]

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        for layer in self.layers.values():
            x = self.linear_forward(*layer, x)
            x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
    
````

````python
from tqdm import tqdm
from functools import partial

ranks = (np.arange(30) + 1) * 10
accs = []
infer_times = np.zeros((len(ranks), 2))

new_dataloader = torch.utils.data.DataLoader(train_loader.dataset, batch_size=1, shuffle=False)
for i, r in tqdm(enumerate(ranks)):
    x_iter = iter(new_dataloader)

    compressed_net = CompressedNet(model, rank=r)
    acc = test(compressed_net, test_loader, verbose=False)
    accs.append(acc)
    with torch.no_grad():
        t = %timeit -o -q -n 100 -r 10 compressed_net(next(x_iter)[0])
    infer_times[i, 0] = t.average
    infer_times[i, 1] = t.stdev

x_iter = iter(new_dataloader)
with torch.no_grad():
    t = %timeit -o -q -n 100 -r 10 model(next(x_iter)[0])
    vanilla_infer_time = (t.average, t.stdev)
vanilla_acc = test(model, test_loader, verbose=False)

fig, axs = plt.subplots(ncols=1, nrows=2, figsize=(10, 13))
axs[0].plot(ranks, accs)
axs[0].set_xlabel('rank')
axs[0].set_ylabel('test acc')
axs[0].set_title('Test accuracy dependence')
axs[0].axhline(y=vanilla_acc, linestyle='--', label='vanilla model accuracy')
axs[0].legend()

axs[1].plot(ranks, infer_times[:, 0])
axs[1].fill_between(ranks, infer_times[:, 0] - infer_times[:, 1], infer_times[:, 0]+infer_times[:, 1], alpha=0.3)
axs[1].set_xlabel('rank')
axs[1].set_ylabel('inference time, s')
axs[1].set_title('Inference time dependence')
axs[1].axhline(y=vanilla_infer_time[0], linestyle='--', label='vanilla model inference time')
axs[1].fill_between(ranks, vanilla_infer_time[0] - vanilla_infer_time[1], vanilla_infer_time[0] + vanilla_infer_time[1], alpha=0.3)
axs[1].legend()

plt.show()
````

Results of test accuracy are explained above - initial model has lots of correlated neurons and may therefore may be "squeezed", "pruned", "distilled", whatever. Indeed the quality of network with low-rank approximations of weight matrices shows same performance when $r>75$. Talking about time: complexity of initial layer: $O(n^2),~n=1000$, complexity with SVD: $O(nr)$ and we can see that linear dependence on $r$ on the plot.

As we have a vector function using the properties of active spaces is not straightforward here (vector of components of $f(x)$ can't be represented as a vectoe of values of some scalar function over the sample $x$).

The derivative of the function with respect to $x$ is:

$$\frac{\partial f(x)}{x} = (W_2 \odot \cos(W_2 cos (W_1x)))~ (W_1 \odot (-\sin(W_1 x))) $$

Here $\odot$ is elementwise product of vector on each column of matrix. 

Sketch of proof:
$$\begin{aligned}&\text{if g is an element-wise function then it's easy to show that}\\&\frac{\partial g(Wx)}{\partial x} = W \odot g'(Wx)\\ &\text{then we apply the chain rule:}\\&\frac{\partial g_2(W_2g_1(W_1x))}{\partial x}= W_2 \odot g_2'(W_2g_1(W_1x)) \frac{\partial g_1(W_1x)}{\partial x} = \\ &= W_2 \odot g_2'(W_2g_1(W_1x)) W_1 \odot g_1'(W_1x)\end{aligned}$$

And therefore if we are lucky enough (with prob $1$) to pick $x$ such that these $\sin$ and $\cos$ do not change the rank of $W_1$ and $W_2$ (which is $r$ almost sure because matrices are random) then diagonal matrix of SVD of the jacobian will have $r$ not zero (actually not close to zero) elements.



````python
import jax.numpy as jnp
import jax
import numpy as np
import random
from jax import grad, jit, vmap, jacfwd, jacrev

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return
    
    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)
    
    def hermione_computes_function(self, x):
        return jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))

    def harry_answers(self):
        key, subkey = jax.random.split(self.key)
        x = jax.random.uniform(subkey, (100, 1), dtype=jnp.float32)
        jac = jax.jacfwd(self.hermione_computes_function)(x).reshape(100, 100)
        u, s, vt = np.linalg.svd(jac)
        r = np.sum(s > 0.1)

        return r
    
    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers()
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````
